```python
from scipy.stats import skew, kurtosis

def generate_full_statistical_summary(dataframe, selected_cols):
    """
    Generates a full statistical summary including describe, skewness, and kurtosis.
    """
    describe_stats = dataframe[selected_cols].describe().T[['mean', '50%', 'std']]
    describe_stats.rename(columns={'50%': 'median'}, inplace=True)
    describe_stats['skewness'] = dataframe[selected_cols].skew()
    describe_stats['kurtosis'] = dataframe[selected_cols].kurt()
    return describe_stats.round(2)

# Columns to analyze
target_columns = [
    'account_balance', 'total_purchases',
    'cash_advance_amt', 'credit_limit_amt',
    'monthly_payments'
]

# Generate and display the full summary
stat_summary_final = generate_full_statistical_summary(df_named, target_columns)
stat_summary_final
```

```python
def plot_tenure_pie(data, filename="plot_tenure_pie.png"):
    """
    Displays a pie chart showing the proportion of customers in each tenure category.
    """
    plt.figure(figsize=(8, 6))
    tenure_counts = data['TENURE'].value_counts().sort_index()
    labels = [f"{int(k)} mo" for k in tenure_counts.index]
    plt.pie(tenure_counts, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title("Customer Distribution by Tenure")
    plt.axis('equal')
    plt.savefig(filename)
    plt.show()
    plt.close()

plot_tenure_pie(df_named)

```

```python
def plot_balance_line(data, filename="plot_balance_line.png"):
    """
    Shows a line plot of average payments across increasing balance groups.
    """
    data_copy = data.copy()
    data_copy['balance_bin'] = pd.qcut(data_copy['account_balance'], q=10, labels=False)
    group_avg = data_copy.groupby('balance_bin')['monthly_payments'].mean()

    plt.figure(figsize=(8, 6))
    plt.plot(group_avg.index, group_avg.values, marker='o')
    plt.xlabel("Balance Quantile Group")
    plt.ylabel("Average Monthly Payment")
    plt.title("Average Monthly Payment by Balance Group")
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    plt.close()

plot_balance_line(df_named)

```

```python
def plot_corr_heatmap(data, filename="plot_corr_heatmap.png"):
    """
    Plots a heatmap of feature correlations.
    """
    plt.figure(figsize=(10, 8))
    corr = data.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title("Correlation Heatmap of Customer Features")
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    plt.close()

plot_corr_heatmap(df_named)
```

```python
# CELL: Elbow Plot Function (Standalone)
def plot_elbow_curve(data, features, max_k=10, filename="elbow_curve_kmeans.png"):
    """
    Plots the elbow curve to help choose the optimal number of clusters (k).
    """
    distortions = []
    for k in range(2, max_k + 1):
        model = KMeans(n_clusters=k, random_state=42)
        model.fit(data[features])
        distortions.append(model.inertia_)

    plt.figure(figsize=(8, 6))
    plt.plot(range(2, max_k + 1), distortions, 'bo-')
    plt.xlabel("Number of Clusters (k)")
    plt.ylabel("Inertia (Distortion)")
    plt.title("Elbow Method for Optimal k")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    plt.close()

# Generate elbow plot
plot_elbow_curve(df_scaled, features_to_scale)
```

```python
# CELL: Cluster Plot with Centroids and New Point Predictions
def plot_kmeans_clusters_with_predictions(data, features, scaler, k=3, filename="kmeans_cluster_plot.png"):
    """
    Performs KMeans clustering and plots customer clusters in 2D using the first two features.
    Also plots new customer predictions.
    """
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    X = data[features]
    labels = kmeans_model.fit_predict(X)
    centers = kmeans_model.cluster_centers_

    # Plot the clusters
    plt.figure(figsize=(9, 7))
    colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']
    for i in range(k):
        plt.scatter(
            X[labels == i].iloc[:, 0], X[labels == i].iloc[:, 1],
            label=f'Cluster {i}', alpha=0.6, color=colors[i % len(colors)]
        )

    # Plot centroids
    plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='X', label='Centroids')

    # New customers for prediction
    new_customers = pd.DataFrame({
        'account_balance': [1500, 3200],
        'total_purchases': [1200, 2500],
        'cash_advance_amt': [0, 500],
        'credit_limit_amt': [3000, 6000],
        'monthly_payments': [800, 1600]
    })

    new_customers_scaled = scaler.transform(new_customers)
    new_labels = kmeans_model.predict(new_customers_scaled)

    for i, point in enumerate(new_customers_scaled):
        plt.scatter(point[0], point[1], marker='D', s=150, edgecolors='black',
                    label=f"New Point {i} (Cluster {new_labels[i]})")

    plt.xlabel(features[0].replace('_', ' ').title())
    plt.ylabel(features[1].replace('_', ' ').title())
    plt.title(f"KMeans Clustering (k={k}) with New Predictions")
    plt.legend(loc='best')
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    plt.close()

# Run the cluster and prediction plot
plot_kmeans_clusters_with_predictions(df_scaled, features_to_scale, final_scaler, k=3)
```

```python
from sklearn.linear_model import LinearRegression

def train_regression_model(dataframe, predictor_col, response_col):
    """
    Trains a linear regression model on the given predictor and response.
    Returns trained model, X, and y.
    """
    X = dataframe[[predictor_col]]
    y = dataframe[response_col]
    model = LinearRegression()
    model.fit(X, y)
    return model, X, y
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_regression_with_ci_and_predictions(model, X, y, predictor_name="X", response_name="y", filename="regression_plot.png"):
    """
    Plots regression line, 95% confidence interval, and predictions with error bars.
    """
    # Original prediction
    y_pred = model.predict(X)

    # Bootstrap for 95% confidence intervals
    n_boot = 1000
    boot_preds = np.zeros((n_boot, len(X)))
    rng = np.random.RandomState(42)

    for i in range(n_boot):
        indices = rng.choice(len(X), size=len(X), replace=True)
        X_sample = X.iloc[indices]
        y_sample = y.iloc[indices]
        boot_model = LinearRegression().fit(X_sample, y_sample)
        boot_preds[i, :] = boot_model.predict(X)

    ci_low = np.percentile(boot_preds, 2.5, axis=0)
    ci_high = np.percentile(boot_preds, 97.5, axis=0)

    # Convert for plotting
    X_vals = X.values.flatten()

    # Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(X_vals, y, alpha=0.4, color='orange', label="Observed Data")
    plt.plot(X_vals, y_pred, color='red', label="Regression Line")
    plt.fill_between(X_vals, ci_low, ci_high, color='pink', alpha=0.4, label="95% Confidence Interval")

    # Predictions on new data
    new_points = pd.DataFrame({predictor_name: [1000, 3000, 6000]})
    new_preds = model.predict(new_points)
    uncertainty = (ci_high - ci_low).mean()

    plt.errorbar(new_points[predictor_name], new_preds, yerr=uncertainty,
                 fmt='o', color='green', capsize=5, label="Predicted Points")

    plt.xlabel(predictor_name.replace("_", " ").title())
    plt.ylabel(response_name.replace("_", " ").title())
    plt.title("Linear Regression with CI and Prediction Uncertainty")
    plt.legend()
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    plt.close()
```

```python
# Example usage
reg_model, X_reg, y_reg = train_regression_model(df_named, 'credit_limit_amt', 'monthly_payments')
plot_regression_with_ci_and_predictions(reg_model, X_reg, y_reg,
                                        predictor_name="credit_limit_amt",
                                        response_name="monthly_payments")
```
